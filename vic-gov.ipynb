```python


from google.transit import gtfs_realtime_pb2
import requests
```

```python
METROTRAIN_TU_URL = "https://wovg-community.gateway.prod.api.vic.gov.au/opendata/v1/gtfsr/metrobus-tripupdates"
METROTRAIN_SA_URL = 'https://wovg-community.gateway.prod.api.vic.gov.au/opendata/v1/gtfsr/metrotrain-servicealerts'
METROTRAIN_VP_URL = "https://wovg-community.gateway.prod.api.vic.gov.au/opendata/v1/gtfsr/metrotrain-vehicleposition-updates"
```

```python
# MBTA
MBTA_SA_URL = "https://cdn.mbta.com/realtime/Alerts.pb"
MBTA_VP_URL = "https://cdn.mbta.com/realtime/VehiclePositions.pb"

feed = gtfs_realtime_pb2.FeedMessage()
response = requests.get(MBTA_VP_URL)
feed.ParseFromString(response.content)
for i, entity in enumerate(feed.entity):
    print(entity)
```
